```python

import os

def replace_special_with_underscores(folder_path):
    if not os.path.exists(folder_path) or not os.path.isdir(folder_path):
        print("Directory not found")
        return
    num_changed = 0
    for root, dirs, files in os.walk(folder_path, topdown=True):
        dirs[:] = [d for d in dirs if d != '.git']
        for filename in files:
            new_filename = filename.replace(' ', '_').replace('-', '_')
            if new_filename != filename:
                old_path = os.path.join(root, filename)
                new_path = os.path.join(root, new_filename)
                os.rename(old_path, new_path)
                print(f"Renamed: {old_path} -> {new_path}")
                num_changed += 1
        new_dirs = []
        for dirname in dirs:
            new_dirname = dirname.replace(' ', '_').replace('-', '_')
            if new_dirname != dirname:
                old_dirpath = os.path.join(root, dirname)
                new_dirpath = os.path.join(root, new_dirname)
                os.rename(old_dirpath, new_dirpath)
                print(f"Renamed: {old_dirpath} -> {new_dirpath}")
                num_changed += 1
                new_dirs.append(new_dirname)
            else:
                new_dirs.append(dirname)
        dirs[:] = new_dirs
    if num_changed == 0:
        print("No files or directories changed")

folder_path = r'C:\Users\tesla\Documents\git\github_reza\rezaaliasgarirenani'
replace_special_with_underscores(folder_path)
```
